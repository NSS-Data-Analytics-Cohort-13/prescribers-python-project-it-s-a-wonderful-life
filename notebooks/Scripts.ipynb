```python
what = '''
SELECT	p.total_claim_count
	,	p.opioid_drug_flag
    ,   p.antibiotic_drug_flag
    ,   z.county
    ,   z.zip
from (SELECT	DISTINCT(P.npi)
        	,	p.nppes_provider_zip5
        	,	pd.total_claim_count
        	,	pd.opioid_drug_flag
        	,	pd.antibiotic_drug_flag
    FROM prescriber AS p
    INNER JOIN (
    				SELECT	p.npi
    					,	p.total_claim_count
    					,	d.opioid_drug_flag
    					,	d.antibiotic_drug_flag
    				FROM prescription as p
    				INNER JOIN drug as d 
    					ON p.drug_name=d.drug_name
    				) AS pd
        ON p.npi=pd.npi) 
        AS p
    inner join (SELECT	z.zip
                	,	f.county
                	,	f.state
                FROM zip_fips as z
                INNER JOIN fips_county as f
                	on z.fipscounty=f.fipscounty
                WHERE f.state = 'TN')
        AS z
on p.nppes_provider_zip5=z.zip;
'''
with engine.connect() as connection:
    tn_opioid = pd.read_sql(text(what), con = connection)
#tn_opioid['npi']=tn_opioid['npi'].astype(int)
tn_opioid.head(10)
```

## Question 2 Who are the top opioid prescibers for the state of Tennessee?

```python
q2 = '''
SELECT	P.npi
	,	CONCAT(p.p_firstname,' ', p.p_lastname) as name
	,	p.total_claim_count
	--,	z.*
from (SELECT	DISTINCT(P.npi)
			,	p.nppes_provider_zip5
			,	p.nppes_provider_last_org_name as p_lastname
			,	p.nppes_provider_first_name as p_firstname
			,	p.nppes_provider_mi as mid
			,	pd.total_claim_count
			,	pd.opioid_drug_flag
			,	pd.antibiotic_drug_flag
		FROM prescriber AS p
		INNER JOIN 
				(
				SELECT	p.npi
					,	p.total_claim_count
					,	d.opioid_drug_flag
					,	d.antibiotic_drug_flag
				FROM prescription as p
				INNER JOIN drug as d 
					ON p.drug_name=d.drug_name
				) AS pd
	ON p.npi=pd.npi) as p
inner join (SELECT	z.zip
				,	f.county
				,	f.state
			FROM zip_fips as z
			INNER JOIN fips_county as f
				on z.fipscounty=f.fipscounty
			WHERE f.state = 'TN') as z
on p.nppes_provider_zip5=z.zip
where opioid_drug_flag = 'Y';
'''
with engine.connect() as connection:
    q2 = pd.read_sql(text(q2), con = connection)
q2['npi']=q2['npi'].astype(int)
q2.head(10)
```

### Q3 Answer

```python
q2.groupby(['npi','name']).sum('total_claim_count').sort_values(by='total_claim_count', ascending=False)
```

### Question 3 What did the trend in overdose deaths due to opioids look like in Tennessee from 2015 to 2018?

```python
q3 = '''
SELECT	p.total_claim_count as opioids_count
	,	z.overdose_deaths
	,	z.year
	,	z.state
from (SELECT	DISTINCT(P.npi)
			,	p.nppes_provider_zip5
			,	p.nppes_provider_last_org_name as p_lastname
			,	p.nppes_provider_first_name as p_firstname
			,	p.nppes_provider_mi as mid
			,	pd.total_claim_count
			,	pd.opioid_drug_flag
			,	pd.antibiotic_drug_flag
		FROM prescriber AS p
		INNER JOIN 
				(
				SELECT	p.npi
					,	p.total_claim_count
					,	d.opioid_drug_flag
					,	d.antibiotic_drug_flag
				FROM prescription as p
				INNER JOIN drug as d 
					ON p.drug_name=d.drug_name
				) AS pd
	ON p.npi=pd.npi) as p
inner join (SELECT	z.zip
				,	f.overdose_deaths
				,	f.year
				,	f.state
			FROM zip_fips as z
			INNER JOIN (select	distinct(f.fipscounty)
							,	d.overdose_deaths
							,	f.state
							,	f.county
							,	d.year	
						from fips_county as f
						inner join overdose_deaths as d
							on f.fipscounty = CAST(d.fipscounty AS VARCHAR(5))
						where d.year between 2015 and 2018
						)as f
				on z.fipscounty=f.fipscounty
			WHERE f.state = 'TN') as z
on p.nppes_provider_zip5=z.zip
where opioid_drug_flag = 'Y';
'''
with engine.connect() as connection:
    q3 = pd.read_sql(text(q3), con = connection)
q3.head(10)
```
