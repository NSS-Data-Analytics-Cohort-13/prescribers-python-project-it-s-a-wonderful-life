## # Tennessee's Opioid Crisis

Opioids are a class of drugs including prescription pain relievers such as oxycodone and hydrocodone, the synthetic opioid fentanyl, and the illegal drug heroin. These drugs produce euphoria in addition to pain relief, which can lead to dependence, addiction, overdose, and death. 

In the late 1990s, opioids began to be prescribed at high rates, which led to widespread misuse and ultimately created a serious national health crisis. In 2019, more than 130 people per day died from opioid-related drug overdoses in the United States. Tennessee has been one of the hardest-hit states in the opioid epidemic, with more than 1300 overdose deaths in 2018.

## Question #1: * Which Tennessee counties had a disproportionately high number of opioid prescriptions?

```python
prescriber['npi']=prescriber['npi'].astype(int)
```

```python
prescription['npi']=prescription['npi'].astype(int)
```

```python
drug_pre = pd.merge(drug, prescription, on='drug_name')
```

```python
pre_pre = pd.merge(drug_pre, prescriber, on='npi')
```

```python
pre_fips = pd.merge(pre_pre, zip_fips, left_on='nppes_provider_zip5', right_on='zip')
```

```python
fips_fips = pd.merge(pre_fips, fips_county, on='fipscounty')
```

```python
fips_pop = pd.merge(fips_fips, population, on='fipscounty')
```

```python
tn_filter = fips_pop[fips_pop['state']=='TN']
```

```python
tn_opioid = tn_filter[tn_filter['opioid_drug_flag']=='Y']
```

```python
tn_opioid_drug_count = fips_pop.groupby('county')['total_claim_count'].sum().reset_index()
```

```python
tn_opioid_drug_count = tn_opioid_drug_count.rename(columns={'total_claim_count': 'total_drug_count'})
```

```python
sum_values = tn_opioid.groupby('county')['total_claim_count'].sum().sort_values(ascending=False).reset_index()
```

```python
tn_opioid_summary = sum_values.merge(tn_opioid_drug_count, on='county')
```

```python
population_county = tn_opioid.groupby('county')['population'].mean().sort_values(ascending=False).reset_index()
```

```python
tn_opioid_summary['percentage'] = tn_opioid_summary['total_claim_count'] / tn_opioid_summary['total_drug_count'] *100
```

```python
tn_opioid_summary = tn_opioid_summary.sort_values(by='percentage', ascending=False).reset_index(drop=True)
```
