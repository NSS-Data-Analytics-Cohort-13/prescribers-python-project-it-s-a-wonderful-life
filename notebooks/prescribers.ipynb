```python
query = '''
SELECT d.drug_name, d.generic_name, d.opioid_drug_flag, d.long_acting_opioid_drug_flag, p.npi, total_claim_count, f.county 
FROM drug AS d
INNER JOIN prescription AS p
ON d.drug_name = p.drug_name
INNER JOIN prescriber as pr
ON pr.npi = p.npi
INNER JOIN zip_fips as z
ON pr.nppes_provider_zip5 = zip
INNER JOIN cbsa as c
ON c.fipscounty = z.fipscounty
INNER JOIN fips_county as f
ON c.fipscounty = f.fipscounty
WHERE  d.opioid_drug_flag = 'Y' AND state ILIKE '%TN'
GROUP BY f.county, d.drug_name, d.generic_name, d.opioid_drug_flag, d.long_acting_opioid_drug_flag, p.npi, total_claim_count;
'''

with engine.connect() as connection:
    opioids = pd.read_sql(text(query), con = connection)

opioids.head(10)
```

```python
opioidcountycount = opioids.groupby('county').sum('total_claim_count').sort_values('county')
opioidcountycount.head(100)                                                                                                                                                                                                                                                              
```

```python
query2 = '''
SELECT pr.npi, pr.nppes_provider_first_name, pr.nppes_provider_last_org_name, d.generic_name, d.opioid_drug_flag, total_claim_count, f.county 
FROM drug AS d
INNER JOIN prescription AS p
ON d.drug_name = p.drug_name
INNER JOIN prescriber as pr
ON pr.npi = p.npi
INNER JOIN zip_fips as z
ON pr.nppes_provider_zip5 = zip
INNER JOIN cbsa as c
ON c.fipscounty = z.fipscounty
INNER JOIN fips_county as f
ON c.fipscounty = f.fipscounty
WHERE  d.opioid_drug_flag = 'Y' AND state ILIKE '%TN'
GROUP BY pr.npi, pr.nppes_provider_first_name, pr.nppes_provider_last_org_name, f.county, d.generic_name, d.opioid_drug_flag, total_claim_count;
'''

with engine.connect() as connection:
    opioidprescribers = pd.read_sql(text(query2), con = connection)
    
opioidprescribers['npi']= opioidprescribers['npi'].astype(int)

opioidprescribers.head(100)

```
