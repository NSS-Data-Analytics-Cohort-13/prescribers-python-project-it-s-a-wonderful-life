```python
# dean's code
what = '''
SELECT	p.total_claim_count
	,	p.opioid_drug_flag
    ,   z.county
from (SELECT	DISTINCT(P.npi)
        	,	p.nppes_provider_zip5
        	,	pd.total_claim_count
        	,	pd.opioid_drug_flag
        	,	pd.antibiotic_drug_flag
    FROM prescriber AS p
    INNER JOIN (
    				SELECT	p.npi
    					,	p.total_claim_count
    					,	d.opioid_drug_flag
    					,	d.antibiotic_drug_flag
    				FROM prescription as p
    				INNER JOIN drug as d
    					ON p.drug_name=d.drug_name
    				) AS pd
        ON p.npi=pd.npi)
        AS p
    inner join (SELECT	z.zip
                	,	f.county
                	,	f.state
                FROM zip_fips as z
                INNER JOIN fips_county as f
                	on z.fipscounty=f.fipscounty
                WHERE f.state = 'TN')
        AS z
on p.nppes_provider_zip5=z.zip;
'''
with engine.connect() as connection:
    tn_opioid = pd.read_sql(text(what), con = connection)
#tn_opioid['npi']=tn_opioid['npi'].astype(int)
tn_opioid.head(10)
```

```python
tn_all = tn_opioid.groupby(['county']).sum('total_claim_count')
```

```python
tn_op = tn_opioid[tn_opioid['opioid_drug_flag'] == 'Y'].groupby(['county']).sum('total_claim_count')
```

```python
tn_op['percentage'] = tn_op/tn_all*100
```

What did the trend in overdose deaths due to opioids look like in Tennessee from 2015 to 2018?

```python
SELECT SUM(overdose_deaths) AS total_deaths, year
FROM overdose_deaths
LEFT JOIN fips_county
ON CAST(overdose_deaths.fipscounty AS TEXT) = fips_county.fipscounty
WHERE state ILIKE '%TN%'
GROUP BY year
ORDER BY year;
```
